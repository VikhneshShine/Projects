# Employee Background and Classification

## Intoduction
No business can exist without employees. They’re the backbone of any successful organization, providing the vital skills and expertise needed to take a business to the next level. Here we are going to explore a dataset which provides various types of information related to employees of a particular company. This project is aimed at studying the various background of different employees and to get an idea about the demographic characterestics and diversity of employees in a company. 

## About the dataset

This dataset contains information about employees in a company, including their educational backgrounds, work history, demographics, and employment-related factors. It has been anonymized to protect privacy while still providing valuable insights into the workforce.
Columns:
- Education_The educational qualifications of employees, including degree, institution, and field of study.
- Joining Year_The year each employee joined the company, indicating their length of service.
- City_The location or city where each employee is based or works.
- Payment Tier_Categorization of employees into different salary tiers.
- Age_The age of each employee, providing demographic insights.
- Gender_Gender identity of employees, promoting diversity analysis.
- Ever Benched_Indicates if an employee has ever been temporarily without assigned work.
- Experience in Current Domain_The number of years of experience employees have in their current field.

# Data Visualisation

## Age Distribution

```python
sns.histplot(data = data, x = 'Age',kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Employees')
plt.show()
```

The histogram shows that employees in the age group 27-29 are the most common.

## City of Origin

```python
sns.countplot(data = data, x ='City')
plt.title('City of origin')
plt.xlabel('City')
plt.ylabel('Number of Employees')
plt.show()
```

The count plot suggests that most of the employees are from Bangalore.

## Educational Qualification

```python
sns.countplot(data = data, x ='Education')
plt.title('Educational Qualification')
plt.xlabel('Education')
plt.ylabel('Number of Employees')
plt.show()
```

The above graph indicates that most of the employees have a Bachelors degree and very few have taken a PHD.

## Gender Distribution

```python
sns.countplot(data = data, x ='Gender')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of Employees')
plt.show()
```

The figure shows that most of employees working for the company are males.

## Year of Joining 

```python
sns.countplot(data = data, x ='JoiningYear')
plt.title('Year of Joining')
plt.xlabel('JoiningYear')
plt.ylabel('Number of Employees')
plt.show()
```

The graph points out that most of the employees joined the company in the year 2017.

## Payment tier Classification

```python
sns.countplot(data = data, x ='PaymentTier')
plt.title('Payment tier')
plt.xlabel('Paymenttier')
plt.ylabel('Number of Employees')
plt.show()
```

The graph shows that most of the employees are tier 3 employees

## Experience Classification

```python
plt.figure(figsize=(10,5))
sns.countplot(data = data, x ='ExperienceInCurrentDomain')
plt.title('Experience Classification')
plt.xlabel('Experience in years')
plt.ylabel('Number of Employees')
plt.show()
```

The figure conveys that most employees have 2 years of experience.

## Age Distribution 

```python
sns.boxplot(data=data,y='Age')
plt.title('Age Distribution Plot')
plt.xlabel('Frequency')
plt.ylabel('Age')
plt.show()
```

The plot shows that average age of the employees in the company lies between 27.5 years and 35 years. There are no visible outliers in the following distribution.

## Experience Distribution

```python
sns.boxplot(data=data,y='ExperienceInCurrentDomain')
plt.title('Experience Distribution Plot')
plt.xlabel('Frequency')
plt.ylabel('Experience in years')
plt.show()
```

The plot shows that average experience of the employees are between 1 to 4 years.

## Gender wise Experience

```python
sns.barplot(data=data, x='Gender', y='ExperienceInCurrentDomain')
plt.title('Experience Distribution Plot')
plt.xlabel('Gender')
plt.ylabel('Experience in years')
plt.show()
```

The following bar graph shows that female employees in the comapany have slightly more experience than male employees. 

## Comparison of Gender and City

```python
sns.countplot(data = data, x ='Gender',hue='City')
plt.title('Comparison of Gender and City')
plt.xlabel('Gender')
plt.ylabel('Number of Employees')
plt.show()
```

The comparison gives us a clear picture that most of the male employees are from Bangalore folloewd by Pune and new Delhi and most of the female employees are from New Delhi followed by Bangalore and Pune.

## Comparison of Gender and Education

```python
sns.countplot(data = data, x ='Gender',hue='Education')
plt.title('Comparison of Gender and Education')
plt.xlabel('Gender')
plt.ylabel('Number of Employees')
plt.show()
```

The following comparison shows that Most of the Male and Female employees have a Bachelors Degree followed by Masters and PHD.

## Comparison of Gender and Joining Year

```python
sns.countplot(data = data, x ='Gender',hue='JoiningYear')
plt.title('Comparison of Gender and Joining Year')
plt.xlabel('Gender')
plt.ylabel('Number of Employees')
plt.show()
```

The following comparison indicates that most of the male and female employees of the company has joined in the year 2017.

## Comparison of Gender and payment tier

```python
sns.countplot(data = data, x ='Gender',hue='PaymentTier')
plt.title('Comparison of Gender and payment tier')
plt.xlabel('Gender')
plt.ylabel('Number of Employees')
plt.show()
```

From this comparison we can infer that most of the male and female employees are tier 3 employees.

## Pie Chart Showing whether the Employees have been benched or not

```python
plt.pie(data=data, x=data.EverBenched.value_counts(),explode = [0.1, 0],labels=['No','Yes'],autopct='%0.2f%%')
plt.title('Pie Chart Showing whether the Employees have been benched or not')
plt.show()
```

The pie chart shows that most of the employees have not been benched.
